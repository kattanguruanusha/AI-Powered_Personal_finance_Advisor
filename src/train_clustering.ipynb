```python
import pandas as pd
import joblib
import os
from sklearn.cluster import KMeans
```

```python
def load_and_clean_data(path):
    df = pd.read_csv("C:\\Users\\katta\\Downloads\\data (1).csv")
    df.dropna(inplace=True)
    return df
```

```python
DATA_PATH = "../data/financial_data.csv"

df = load_and_clean_data(DATA_PATH)
df.columns
```

```python
def create_features(df):
    expense_columns = [
        "Rent", "Loan_Repayment", "Insurance", "Groceries", "Transport",
        "Eating_Out", "Entertainment", "Utilities", "Healthcare",
        "Education", "Miscellaneous"
    ]
    
    df["Total_Expenses"] = df[expense_columns].sum(axis=1)
    
    df["expense_ratio"] = df["Total_Expenses"] / df["Income"]
    df["savings_ratio"] = df["Desired_Savings"] / df["Income"]
    
    return df
```

```python
df = create_features(df)

df[["Income", "Total_Expenses", "Desired_Savings", "expense_ratio", "savings_ratio"]].head()
```

```python
X = df[["expense_ratio", "savings_ratio"]]

kmeans = KMeans(n_clusters=3, random_state=42)
df["spending_cluster"] = kmeans.fit_predict(X)

df["spending_cluster"].value_counts()
```

```python
os.makedirs("../outputs/models", exist_ok=True)

joblib.dump(kmeans, "../outputs/models/customer_segmentation_model.pkl")

print("✅ Clustering model saved successfully")
```
