```python
def generate_advice(savings_ratio: float, cluster: int):
    """
    Generates personalized financial advice based on
    savings behavior and spending cluster.

    Parameters:
    savings_ratio (float): savings / monthly_income
    cluster (int): spending cluster from KMeans

    Returns:
    list: financial advice strings
    """

    advice = []

    # Savings-based advice
    if savings_ratio >= 0.30:
        advice.append(
            "Excellent savings habit. Consider long-term investments such as mutual funds or fixed deposits."
        )
    elif savings_ratio >= 0.15:
        advice.append(
            "Good savings behavior. Try optimizing discretionary expenses to improve savings further."
        )
    else:
        advice.append(
            "Low savings detected. Create a strict monthly budget and prioritize essential expenses."
        )

    # Cluster-based advice
    if cluster == 0:
        advice.append(
            "You belong to the overspending group. Focus on expense tracking and spending discipline."
        )
    elif cluster == 1:
        advice.append(
            "You demonstrate balanced spending habits. Maintain consistency and review finances monthly."
        )
    elif cluster == 2:
        advice.append(
            "You are a high saver. Explore wealth growth strategies such as equity investments."
        )

    return advice
```

```python
# Example test case
sample_savings_ratio = 0.22
sample_cluster = 1

recommendations = generate_advice(sample_savings_ratio, sample_cluster)

for idx, rec in enumerate(recommendations, start=1):
    print(f"{idx}. {rec}")
```
