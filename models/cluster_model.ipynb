```python
import pandas as pd
import pickle
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import os
```

```python
df = pd.read_csv("C:\\Users\\katta\\Downloads\\data (1).csv")
df.head()
```

```python
cluster_features = [
    "Income",
    "Rent",
    "Loan_Repayment",
    "Groceries",
    "Transport",
    "Eating_Out",
    "Entertainment",
    "Utilities",
    "Healthcare",
    "Education",
    "Miscellaneous"
]

X = df[cluster_features]
```

```python
# scale feature
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# train KMeans model
kmeans = KMeans(
    n_clusters=4,
    random_state=42,
    n_init=10
)

df["user_cluster"] = kmeans.fit_predict(X_scaled)
```

```python
# Save cluster model feature names
with open("../models/cluster_model_columns.pkl", "wb") as f:
    pickle.dump(X.columns.tolist(), f)
```

```python
os.makedirs("../models", exist_ok=True)

with open("../models/cluster_model.pkl", "wb") as f:
    pickle.dump(kmeans, f)

with open("../models/cluster_scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)

print("✅ Cluster model & scaler saved successfully")
```
